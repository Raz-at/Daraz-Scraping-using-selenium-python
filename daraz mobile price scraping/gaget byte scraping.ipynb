```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from time import sleep
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import csv
import pandas as pd
from selenium.common.exceptions import NoSuchElementException
import re
```

```python
# Set up the driver
path = "C:\Program Files\Google\Chrome\chromedriver.exe"
service = Service(executable_path=path)
driver = webdriver.Chrome(service=service)
driver.get("https://www.gadgetbytenepal.com/category/mobile-price-in-nepal/")
driver.maximize_window()
```

```python
phone_data = []

phone_elements = driver.find_elements(By.XPATH, '//div[@class="su-table su-table-alternate"]/table/tbody/tr/td[1]')
price_elements = driver.find_elements(By.XPATH, '//div[@class="su-table su-table-alternate"]/table/tbody/tr/td[2]')

for phone_element, price_element in zip(phone_elements, price_elements):
    phone_name = phone_element.text
    phone_price = price_element.text
    
    matches = re.findall(r'\d+\s*(?:GB|TB)', phone_name)
    phone_storage = matches[0] if matches else "N/A"
   
    matches = re.findall(r'(\d+)\s*[+/]', phone_name)
    phone_ram = matches[0]+"GB" if matches else "N/A"

    
    phone_info = {"Name": phone_name, "Price": phone_price, "Storage": phone_storage , "RAM": phone_ram}
    phone_data.append(phone_info)
```

```python
print(len(phone_data))
```

```python
for phone in phone_data:
        print("Name:", phone["Name"],
        print("Price:", phone["Price"])
        print("Storage:", phone["Storage"])
        print("RAM:", phone["RAM"])
        print()
```

```python
value = "iphone 14 pro"
value = value.lower()
value = value.replace("samsung ", "")
value = value.replace("apple ", "")
value_words = value.split()
index = value_words.index(value_words[-1])

found_phones = []

for phone in phone_data:
    words = phone["Name"].split()
    tot_words = ' '.join(words[:index+1])
    if tot_words.lower() == value:
        found_phones.append(phone)

if found_phones:
    for found_phone in found_phones:
        name = re.split(r' [(]', found_phone["Name"], 1)[0]
        print("Name:", name)
        print("Price:", found_phone["Price"])
        print("Storage:", found_phone["Storage"])
        print("RAM:", found_phone["RAM"])
        print()
else:
    print("No phones found with the given value.")

```
